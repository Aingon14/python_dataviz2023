# 1. ในแต่ละจังหวัด ส.ส.แบ่งเขต พรรคใดมีคะแนนเสียงมากที่สุด

```python
# คะแนนเสียงส.ส.เขต
result_constituencies_Candidate.head()
```

```python
# ผู้สมัครส.ส.
Candidate_constituency.head()
```

```python
# ชื่อพรรคที่สังกัด
info_party_overview.head()
```

```python
# merge คะแนนเสียงส.ส.เขต x ผู้สมัครส.ส.
data_table_merge1 = result_constituencies_Candidate.merge(Candidate_constituency, how ='left', left_on='mp_app_id',right_on='mp_app_id')
print(data_table_merge1.shape)
data_table_merge1.head()
```

```python
# merge คะแนนเสียงส.ส.เขต x ผู้สมัครส.ส. x ชื่อพรรคที่สังกัด
data_table_merge2 = data_table_merge1.merge(info_party_overview, how ='left', left_on='party_id',right_on='id')
print(data_table_merge2.shape)
data_table_merge2.head()
```

```python
# merge คะแนนเสียงส.ส.เขต x ผู้สมัครส.ส. x ชื่อพรรคที่สังกัด x จำนวนผู้มาใช้สิทธิ์
data_table_merge3 = data_table_merge2.merge(result_constituencies_status, how ='left', left_on='cons_id',right_on='cons_id')
print(data_table_merge3.shape)
data_table_merge3.head()
```

```python
# merge คะแนนเสียงส.ส.เขต x ผู้สมัครส.ส. x ชื่อพรรคที่สังกัด x จำนวนผู้มาใช้สิทธิ์ x จังหวัด
data_table_merge4 = data_table_merge3.merge(info_province, how ='left', left_on='prov_id',right_on='prov_id')
print(data_table_merge4 .shape)
data_table_merge4 .head()
```

```python
# merge คะแนนเสียงส.ส.เขต x ผู้สมัครส.ส. x ชื่อพรรคที่สังกัด x จำนวนผู้มาใช้สิทธิ์ x จังหวัด x ภาค
data_table_merge5 = data_table_merge4.merge(Region, how ='left', left_on='province_id',right_on='province_id')
print(data_table_merge5.shape)
data_table_merge5.head()
```
